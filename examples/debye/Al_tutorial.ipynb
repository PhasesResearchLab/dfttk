# Import neccessary libraries

```python
%load_ext autoreload
%autoreload 2

# Third-party imports
import numpy as np

# DFTTK importscd
from dfttk.debye import DebyeGruneisen
```

# DebyeGruneisen

```python
# Parameters for the Debye-Gruneisen model
# Atomic mass, V0, B, and BP come from Al PBE calculations
number_of_atoms = 4
volumes = np.linspace(0.98 * 60, 1.02 * 74, 10)
temperatures = np.arange(0, 1010, 10)
atomic_mass = 26.981
V0 = 66.10191547034127
B = 77.92792067011315
BP = 4.612739661291564
scaling_factor = 0.617
gruneisen_x = 2 / 3
```

```python
# Process the Debye-Gruneisen model to compute properties as a function of temperature and volume
debye = DebyeGruneisen()
debye.process(
    number_of_atoms,
    volumes,
    temperatures,
    atomic_mass,
    V0,
    B,
    BP,
    scaling_factor,
    gruneisen_x,
)
```

```python
# Plot the results
# Options for type: 'helmholtz_energy_vs_temperature', 'entropy_vs_temperature', 'heat_capacity_vs_temperature', 
# 'helmholtz_energy_vs_volume', 'entropy_vs_volume', 'heat_capacity_vs_volume'
fig = debye.plot_vt("helmholtz_energy_vs_temperature")
fig.show()
```

```python
# When plotting vs. volume, you can specify particular temperatures to plot
selected_temperatures = np.array([0, 300, 600, 900])
debye.plot_vt('helmholtz_energy_vs_volume', selected_temperatures=selected_temperatures)
```

```python
# When plotting vs. temperature, you can specify particular volumes to plot
selected_volumes = np.array([60, 65, 70, 75])
debye.plot_vt('helmholtz_energy_vs_temperature', selected_volumes=selected_volumes)
```
